```python
# 1 area and perimeter of circle and rectangle using module
import circle
from rect import*
while(1):
    print("\n 1.circle \n 2.rectangle\n 3.exit")
    ch=int(input("enter any option:"))
    if ch==1:
        r=int(input("enter radius:"))
        carea=circle.circlearea(r)
        cper=circle.circleperimeter(r)
        print("area of circle is ",carea,"\n perimeter of circle is ",cper)
    elif ch==2:
            a=int(input("enter length:"))
            b=int(input("enter breadth:"))
            rarea=rectarea(a,b)
            rper=rectperimeter(a,b)
            print("area of rectangle is ",rarea,"\n perimeter of rectangle is ",rper)
    else:
        print("exiting.....")
        break
```

```python
# 2 area and perimeter of circle,rectangle,sphere and cuboid using module

from graphics.rect import*
from graphics.circle import*
from graphics.graphics3d import cuboid
from graphics.graphics3d import sphere
while(1):
    print("\n1.circle\n2.rectangle\n3.sphere\n4cuboid \n5.exit")
    ch=int(input("enter any option:"))
    if ch==1: 
        r=int(input("enter radius:"))
        carea=circlearea(r) 
        cper=circleperimeter(r) 
        print("area of circle is ",carea,"\n perimeter of circle is ",cper)
    elif ch==2:
        a=int(input("enter length:"))
        b=int(input("enter breadth:"))
        rarea=rectarea(a,b)
        rper=rectperimeter(a,b)
        print("area of rectangle is ",rarea,"\n perimeter of rectangle is ",rper)
    elif ch==3:
        r=int(input("enter radius:"))
        sarea=sphere.sphere_area(r)
        sper=sphere.circumference(r)
        print("area of sphere is ",sarea,"\n perimeter of sphere is ",sper)

    elif ch==4:
        l=int(input("enter length:"))
        b=int(input("enter breadth:"))
        h=int(input("enter height:"))
        carea=cuboid.cuboid_area(l,b,h)
        cper=cuboid.cuboid_perimeter(l,b,h)
        print("area of cuboid is ",carea,"\n perimeter of cuboid is ",cper)        
    else:
        print("exiting.....")
        break
```
